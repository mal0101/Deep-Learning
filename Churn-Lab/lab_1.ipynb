```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("data/Churn_Modelling_.csv")
target = df.pop('Exited')
df.head()
```

```python
X = df.drop(columns=["RowNumber", "CustomerId", "Surname"])
```

```python
X['Gender'] = X['Gender'].map({'Female': 0, 'Male': 1})
```

```python
X = pd.get_dummies(X, columns=['Geography'], drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann = Sequential()
ann.add(Dense(units=6, activation='leaky_relu', input_dim=X_train.shape[1]))
ann.add(Dense(units=5, activation='leaky_relu'))
ann.add(Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size=32, epochs=100)
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```
